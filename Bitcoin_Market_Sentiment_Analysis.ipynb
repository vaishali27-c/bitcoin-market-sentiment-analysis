# **Exploratory Data Analysis (EDA)**

```python
trade_count = merged_df['classification'].value_counts()

print(trade_count)
```

## Number of Trades by Market Sentiment

```python
from matplotlib import colors
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid", context="talk")

plt.figure(figsize=(10,6))

palette = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA726', '#7E57C2']

ax = sns.barplot(
    x=trade_count.index,
    y=trade_count.values,
    palette=palette
)

plt.title("Number of Trades by Market Sentiment",
          fontsize=18,
          fontweight='bold')

plt.xlabel("Market Sentiment", fontsize=13)
plt.ylabel("Number of Trades", fontsize=13)

plt.xticks(rotation=25)

for i, value in enumerate(trade_count.values):
    ax.text(i, value+800,
            f"{value:,}",
            ha='center',
            fontsize=11,
            fontweight='bold')

sns.despine()

plt.tight_layout()

plt.show()
```

# Insight

## - Fear recorded the highest trading activity with 61,837 trades.
## - Extreme Fear had the lowest trading activity.
## - Trading frequency changes noticeably across different market sentiments.

```python
avg_pnl = merged_df.groupby('classification')['Closed PnL'].mean().sort_values(ascending=False)

print(avg_pnl)
```

## Average Closed PnL by Market Sentiment

```python
plt.figure(figsize=(10,6))

palette = sns.color_palette("rocket", len(avg_pnl))

ax = sns.barplot(
    y=avg_pnl.index,
    x=avg_pnl.values,
    palette=palette
)

plt.title("Average Closed PnL by Market Sentiment",
          fontsize=18,
          fontweight='bold')

plt.xlabel("Average Closed PnL (USD)")
plt.ylabel("")

for i, value in enumerate(avg_pnl.values):
    ax.text(value+0.5,
            i,
            f"{value:.2f}",
            va='center',
            fontsize=11,
            fontweight='bold')

sns.despine()

plt.tight_layout()

plt.show()
```

## Summary

- **Trading Activity:** 'Fear' sentiment recorded the highest number of trades, followed by 'Greed'. Trading activity generally increased towards late 2024 and early 2025.
- **Profitability (Closed PnL):** 'Extreme Greed' showed the highest average `Closed PnL`, with 'Fear' coming in second. Most trades resulted in small profits or losses, but significant outliers (both positive and negative) were present across all sentiments.
- **Win Rate:** 'Extreme Greed' had the highest win rate, while 'Extreme Fear' had the lowest.
- **Trade Size vs. PnL:** No clear linear correlation was observed between trade size and `Closed PnL`, suggesting other factors heavily influence profitability.
- **Asset Performance:** Certain coins, like '@109', 'AVAX', and 'ENA', demonstrated higher average `Closed PnL`.
- **Buy vs. Sell:** Buy and sell trades remained relatively balanced across different market sentiments, though sell trades slightly outnumbered buy trades during 'Fear' and 'Greed'.
- **Correlations:** Most numerical features had weak correlations with `Closed PnL`, except for `Trade Size` and `Fee` which were strongly positively correlated.

## Recommendations

- **Capitalize on 'Extreme Greed' and 'Fear':** Focus trading strategies during 'Extreme Greed' and 'Fear' sentiments, as these periods showed the highest average `Closed PnL` and win rates.
- **Strategic Asset Selection:** Prioritize trading assets that have historically shown higher average `Closed PnL`, such as '@109', 'AVAX', and 'ENA'.
- **Risk Management for Outliers:** Implement robust risk management strategies to mitigate potential losses from the significant negative outliers observed across all sentiments.
- **Diversify Strategy:** Given the lack of strong correlation between trade size and PnL, and the general weak correlations among numerical features, traders should focus on diverse strategies rather than relying on single indicators.
- **Monitor Fee Impact:** Since `Trade Size` and `Fee` are positively correlated, consider strategies to optimize fees, especially for larger trades, to improve net profitability.
